```python
'''
convenience functions
march 2105
'''

%matplotlib inline
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

# numpy output options
np.set_printoptions(precision=2,suppress=True)

pi=np.pi
j=1j
```

```python
def clean(x):
    tol=1e-10
    y=np.copy(x)
    y[abs(y) < tol] = 0.0
    return y
```

```python
def myplotstyle(**kwargs):
    'customize plot style'
    margins=kwargs.pop('margins',0.1)
    
    plt.margins(margins)
    plt.axhline(color='k')
    plt.axvline(color='k')
    plt.grid(True,which='both')
    plt.legend()
```

```python
def complexplot(*args,**kwargs):
    '''
    plot complex sequences
    arguments: (range, values, style, title='', xlabel='')
    '''
    plot_range=args[0]
    values=args[1]
    style=args[2]
    title=kwargs.pop('title','')
    xlabel=kwargs.pop('xlabel','')
    
    plt.figure(1,figsize=(16,4))#dpi=120 has no effect 
    plt.subplots_adjust(wspace=.15,hspace=0.03)
    plt.suptitle(title)
    
    plt.subplot(2,2,1)
    cleaned=clean(np.abs(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('magnitude')
    plt.gca().axes.get_xaxis().set_ticklabels([])#clear ticks from xlabel
    myplotstyle()
    
    plt.subplot(2,2,3)
    cleaned=clean(np.angle(values))
    cleaned[cleaned>pi-.01]=-pi#choose -pi over +pi
    plt.plot(plot_range,cleaned/pi,style,**kwargs)
    plt.ylabel('phase/$\pi$')
    myplotstyle()
    plt.xlabel(xlabel)
    
    plt.subplot(2,2,2)
    cleaned=clean(np.real(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('real')
    plt.gca().axes.get_xaxis().set_ticklabels([])#clear ticks from xlabel
    myplotstyle()
    
    plt.subplot(2,2,4)
    cleaned=clean(np.imag(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('imaginary')
    myplotstyle()
    plt.xlabel(xlabel)
```

```python
def logplot(*args,**kwargs):
    '''
    plot in log scale
    arguments: (range, mag, phase, title='', margins=0.1, xlabel='frequency (HZ)')
    '''
    plot_range=args[0]
    mag=args[1]
    phase=args[2]
    title=kwargs.pop('title','')
    margins=kwargs.pop('margins',0.1)
    xlabel=kwargs.pop('xlabel','frequency (Hz)')
    
    plt.figure(1,figsize=(16,4))#dpi=120 has no effect 
    plt.subplots_adjust(wspace=.15,hspace=0.03)
    plt.suptitle(title)
    
    plt.subplot(2,1,1)
    plt.loglog(plot_range, mag,lw=2,**kwargs)
    plt.ylabel('magnitude')
    plt.gca().axes.get_xaxis().set_ticklabels([])#clear ticks from xlabel
    myplotstyle(margins=margins)
    
    plt.subplot(2,1,2)
    plt.semilogx(plot_range, phase/180,lw=2,**kwargs)
    plt.ylabel('phase/$\pi$')
    myplotstyle(margins=margins)
    plt.xlabel(xlabel)
```

```python
def timeplot(*args,**kwargs):
    '''
    plot in time domain
    arguments: (range, values, title='', ylabel='', xlabel='time (s)')
    '''
    plot_range=args[0]
    values=args[1]
    title=kwargs.pop('title','')
    ylabel=kwargs.pop('ylabel','')
    xlabel=kwargs.pop('xlabel','time (s)')
    
    plt.figure(1,figsize=(16,4))#dpi=120 has no effect 
    plt.title(title)
    plt.plot(plot_range,values,lw=2,**kwargs)
    plt.ylabel(ylabel)
    myplotstyle()
    plt.xlabel(xlabel)
```

```python
def pzmap(*args,**kwargs):
    '''
    pole-zero map
    arguments: (system, title='pole-zero map', ylabel='Im{s}', xlabel='Re{s}')
    '''
    sys=args[0]
    title=kwargs.pop('title','pole-zero map')
    ylabel=kwargs.pop('ylabel','Im{s}')
    xlabel=kwargs.pop('xlabel','Re{s}')
    axis=kwargs.pop('axis','equal')
    
    zeros=np.roots(sys.num)
    poles=np.roots(sys.den)
    
    plt.figure(1,figsize=(6,6))#dpi=120 has no effect 
    plt.title(title)
    plt.plot(np.real(zeros),np.imag(zeros),'o',ms=12,mew=2,mfc='none',mec='b',**kwargs)
    plt.plot(np.real(poles),np.imag(poles),'bx',ms=12,mew=2,**kwargs)
    plt.ylabel(ylabel)
    myplotstyle()
    plt.axis(axis)
    plt.xlabel(xlabel)
```

```python
def combinedplot(*args,**kwargs):
    '''
    Combined system plot: Bode, step, pzmap
    arguments: (system)
    '''
    sys=args[0]
    axis=kwargs.pop('axis','equal')
    #t=kwargs.pop('t','none')#TODO: pass array of time for the step function
    plt.figure(1,figsize=(16,8))#dpi=120 has no effect
    plt.subplots_adjust(wspace=.2,hspace=.5)
    
    w, Hs = signal.freqresp(sys)
    freqs=w/(2*pi)
    mag=np.abs(Hs)
    phase=np.angle(Hs)*180/pi
    margins=0.4

    plt.subplot(4,2,1)
    plt.title('Bode plot')
    plt.loglog(w, mag,lw=2,**kwargs)
    plt.ylabel('magnitude')
    plt.gca().axes.get_xaxis().set_ticklabels([])#clear ticks from xlabel
    myplotstyle(margins=margins)
    
    plt.subplot(4,2,3)
    plt.semilogx(w, phase/180,lw=2,**kwargs)
    plt.ylabel('phase/$\pi$')
    myplotstyle(margins=margins)
    plt.xlabel('$w$ (rad/s)')
    
    t,s=signal.step(sys)#TODO: pass argument T=<array>
    
    plt.subplot(2,2,3)
    plt.title('step response')
    plt.plot(t,s,lw=2,**kwargs)
    myplotstyle()
    plt.xlabel('time (s)')
    
    zeros=np.roots(sys.num)
    poles=np.roots(sys.den)
       
    plt.subplot(1,2,2)
    plt.title('pole-zero map')
    plt.plot(np.real(zeros),np.imag(zeros),'o',ms=12,mew=2,mfc='none',mec='b',**kwargs)
    plt.plot(np.real(poles),np.imag(poles),'bx',ms=12,mew=2,**kwargs)
    plt.ylabel('Im{s}')
    myplotstyle()
    plt.axis(axis)
    plt.xlabel('Re{s}')
```
